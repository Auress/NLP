```python
import numpy
import torch
import re
```

```python
from transformers import GPT2LMHeadModel
from transformers import GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("sberbank-ai/rugpt3large_based_on_gpt2")
model = GPT2LMHeadModel.from_pretrained("sberbank-ai/rugpt3large_based_on_gpt2", output_attentions=True)
```

```python
# device = "cuda" if torch.cuda.is_available() else "cpu"
device = "cpu"
model.to(device)
```

```python
#inputs = tokenizer("Как дела", return_tensors="tf")
#outputs = model(**inputs)
```

```python
start = tokenizer.encode(
    "как обучить нейронную сеть", 
    return_tensors="pt"
)
```

```python
result = model.generate(
    start.to(device), 
    max_length=50,
    num_beams=10,
    early_stopping=True, 
    no_repeat_ngram_size=3,
)
```

```python
txt = (tokenizer.decode(result.cpu().flatten().numpy(), skip_special_tokens=True))
```
