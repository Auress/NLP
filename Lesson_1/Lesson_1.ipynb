```python
# 2
def del_user(text):
    to_del_list = re.findall("@[\w]*", text)
    for to_del in to_del_list:
        text = re.sub(to_del, " ", text)
    return text
```

```python
data['clean_tweet'] = data['clean_tweet'].apply(lambda x: del_user(x))
```

```python
def change_with_dict(text, dictionary):
    for word in text.split():
        if word in dictionary.keys():
            text = re.sub(re.escape(word), dictionary[word], text)
    
    return text
```

```python
# 11
data['tweet_token'] = data['clean_tweet'].apply(lambda x: nltk.tokenize.word_tokenize(x))
```

```python
# 12
stop_words = set(nltk.corpus.stopwords.words("english"))
data['tweet_token_filtered'] = data['tweet_token'].apply(lambda x: [word for word in x if not word in stop_words])
```

```python
# 13
Stemmer = nltk.stem.PorterStemmer()
data['tweet_stemmed'] = data['tweet_token_filtered'].apply(lambda x: [Stemmer.stem(word) for word in x])
```

```python
# 14
Lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()
data['tweet_lemmatized'] = data['tweet_token_filtered'].apply(lambda x: [Lemmatizer.lemmatize(word) for word in x])
```
